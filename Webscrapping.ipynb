```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url ='https://en.wikipedia.org/wiki/List_of_South_African_films#2020s'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table',class_ ='wikitable')
```

```python
table = soup.find_all('table')[9]
```

```python
table
```

```python
world_table_titles = table.find_all('th')

world_table_titles
```

```python
world_table_titles = [title.text.strip() for title in world_table_titles]

world_table_titles
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
        if len(individual_row_data) ==6:
        length = len(df)
        df.loc[length] = individual_row_data
```
